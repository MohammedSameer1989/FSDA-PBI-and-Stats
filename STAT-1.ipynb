Q1- Plot a histogram,
10, 13, 18, 22, 27, 32, 38, 40, 45, 51, 56, 57, 88, 90, 92, 94, 99


```python
Ans-To plot a histogram in Python, you can use libraries like Matplotlib. Below is a Python code to plot a histogram for the given data using Matplotlib:
```

```python
import matplotlib.pyplot as plt

data = [10, 13, 18, 22, 27, 32, 38, 40, 45, 51, 56, 57, 88, 90, 92, 94, 99]

plt.hist(data, bins=10, color='skyblue', edgecolor='black')

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')

plt.show()
```

```python
This code will create a histogram for the given data. The data variable contains the values for which we want to create the histogram. We use plt.hist() function to plot the histogram, specifying the number of bins using the bins parameter. Then, we add labels to the x-axis and y-axis using plt.xlabel() and plt.ylabel() functions respectively, and finally, we give a title to the plot using plt.title() function. Finally, we display the plot using plt.show().
```

Q5-In left & right-skewed data, what is the relationship between mean, median & mode?
Draw the graph to represent the same.


```python
Ans-In left-skewed data (negative skewness), the mean is less than the median, which is less than the mode. Conversely, in right-skewed data (positive skewness), the mean is greater than the median, which is greater than the mode.

Here's how the relationship between mean, median, and mode looks like visually on a graph:

For left-skewed data:
```

```python
         /          
        /            
------/--------\------
      ^        ^
      |        |
     Mode   Median
             |
           Mean
```

```python
For right-skewed data:
             \          
          \            
-------\--------/------
       ^        ^
       |        |
      Mode    Median
              |
            Mean
In a left-skewed distribution, the mean is pulled to the left by the long tail of low values, making it less than the median. The mode is the highest point and is to the right of the mean and median.

In a right-skewed distribution, the mean is pulled to the right by the long tail of high values, making it greater than the median. The mode is still the highest point but is to the left of the mean and median.
```
